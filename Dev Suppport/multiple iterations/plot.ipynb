```python
import pandas as pd
import matplotlib.pyplot as plt

# After appending your results, create the DataFrame
df = pd.read_csv('quantum_network_metrics_refined.csv')
```

```python
# Loop through each unique combination of crosstalk_strength and crosstalk_type
unique_combinations = df[['crosstalk_strength', 'crosstalk_type']].drop_duplicates()

for _, row in unique_combinations.iterrows():
    crosstalk_strength = row['crosstalk_strength']
    crosstalk_type = row['crosstalk_type']
    
    # Filter the DataFrame based on crosstalk_strength and crosstalk_type
    filtered_df = df[(df['crosstalk_strength'] == crosstalk_strength) & (df['crosstalk_type'] == crosstalk_type)]
    
    # Plot the filtered data
    plt.figure(figsize=(10, 6))
    plt.plot(filtered_df['iterations'], filtered_df['effective_throughput_qubits_per_sec'], marker='o', linestyle='-', label=f'{crosstalk_type} - {crosstalk_strength}')
    
    # Add labels and title
    plt.title(f"Effective Throughput vs Iterations ({crosstalk_type} - {crosstalk_strength})", fontsize=14)
    plt.xlabel("Iterations", fontsize=12)
    plt.ylabel("Effective Throughput (qubits/sec)", fontsize=12)
    plt.grid(True)
    plt.legend()
    
    # Show the plot
    plt.show()
```
